```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
##drop month,day and yyear
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Encoding
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.tail()
```

```python
df['Classes'].value_counts()
```

```python
## Independent And dependent features
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
X.head()
```

```python
y
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature Selection based on correlaltion
X_train.corr()
```

```python
## Check for multicollinearity
plt.figure(figsize=(10,10))
corr=X_train.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
def correlation(dataset, threshold): #function to find correlated features 
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features=correlation(X_train,0.85)
```

```python
## drop features when correlation is more than 0.85 
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

## Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train) #fit and transform because we have to learn the parameters from train data
X_test_scaled=scaler.transform(X_test) # only transform because we have to use the same parameters learned from train data
```

```python
X_train_scaled
```

## Box Plots To understand Effect Of Standard Scaler

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
plt.show()
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression   # Import Linear Regression model from sklearn
from sklearn.metrics import mean_absolute_error      # Import MAE metric to measure prediction error
from sklearn.metrics import r2_score                 # Import R2 score to measure model performance

linreg = LinearRegression()                          # Create a Linear Regression model object

linreg.fit(X_train_scaled, y_train)                  # Train (fit) the model using scaled training features and labels

y_pred = linreg.predict(X_test_scaled)               # Predict the output for scaled test features

mae = mean_absolute_error(y_test, y_pred)            # Calculate MAE  average difference between actual and predicted values
score = r2_score(y_test, y_pred)                     # Calculate R  how well the model explains the variance in output

print("Mean absolute error", mae)                    # Print MAE value
print("R2 Score", score)                             # Print R score value

plt.scatter(y_test, y_pred)                          # Scatter plot of actual vs predicted values
plt.xlabel("Actual FWI")                             # Label for X-axis
plt.ylabel("Predicted FWI")                          # Label for Y-axis
plt.title("Actual vs Predicted FWI")                 # Plot title
plt.show()                                           # Display the plot
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso=Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

### Cross Validation Lasso

```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scaled,y_train)
```

```python
lassocv.alpha_
```

```python
lassocv.alphas_
```

```python
lassocv.mse_path_ #mean squared error for each fold and each alpha value
```

```python
y_pred=lassocv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

## Ridge Regression model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train)
y_pred=ridgecv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

```python
ridgecv.get_params() #get the parameters of the model
```

## Elasticnet Regression

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
elastic=ElasticNet()
elastic.fit(X_train_scaled,y_train)
y_pred=elastic.predict(X_test_scaled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
```

```python
from sklearn.linear_model import ElasticNetCV
elasticcv=ElasticNetCV(cv=5)
elasticcv.fit(X_train_scaled,y_train)
y_pred=elasticcv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

```python
elasticcv.alphas_
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

models = {
    "Ridge": ridgecv,
    "Lasso": lassocv,
    "ElasticNet": elasticcv
}

for name, model in models.items():
    y_pred = model.predict(X_test_scaled)

    print(f"\n===== {name} Results =====")
    print("Best alpha:", model.alpha_)

    if name == "ElasticNet":
        print("L1 Ratio:", model.l1_ratio_)

    print("R2 Score:", r2_score(y_test, y_pred))
    print("MAE:", mean_absolute_error(y_test, y_pred))
    print("MSE:", mean_squared_error(y_test, y_pred))

    print("Coefficients:", model.coef_)
```

```python
for name, model in models.items():
    y_pred = model.predict(X_test_scaled)
    
    plt.figure(figsize=(5,5))
    plt.scatter(y_test, y_pred)
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title(f"{name} CV: Actual vs Predicted")
    plt.show()
```

```python
plt.scatter(y_test, y_pred, color='blue', label='Predicted')
plt.scatter(y_test, y_test, color='red', label='Actual')

plt.xlabel("Actual FWI")
plt.ylabel("Predicted FWI")
plt.title("Actual vs Predicted FWI")
plt.legend()
plt.show()

```
